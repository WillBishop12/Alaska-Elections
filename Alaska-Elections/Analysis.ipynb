```python
import pandas as pd
import numpy as np
```

```python
ElectionsDF = pd.read_csv("elections_2014_2016.csv", index_col=0, header=[0,1,2])
ElectionsDF
```

```python
#Use only statewide elections, i.e. those in which all precincts voted
ElectionsDF.dropna(axis=1, inplace=True)
ElectionsDF
```

```python
import sklearn.cluster
```

```python
# k-means clustering with 2 to 10 clusters
kmeans = {}
results = {}
for k in range(2, 11):
    kmeans[k] = sklearn.cluster.KMeans(n_clusters=k)
    kmeans[k].fit(ElectionsDF)
    results[k] = kmeans[k].predict(ElectionsDF)
    print "%0d: score %f" %(k, kmeans[k].score(ElectionsDF))
```

```python
results[10]
```

```python
results[5]
```

```python
#Both 10-cluster and 5-cluster cases assigned almost all "absentee/question" precincts to their own cluster - not good.
#This is probably because turnout numbers aren't meaningful for those precincts - represent percentage of total in district.
#Rerun in two ways: first eliminating absentee/question ballots so we can use turnout, then without using turnout numbers.
```

```python
ElectionsDF_NoAbs = ElectionsDF[ElectionsDF.index.str.contains("District") == False]
ElectionsDF_NoAbs
```

```python
# k-means clustering with 2 to 10 clusters - NO absentee/question precincts
kmeans = {}
results = {}
for k in range(2, 11):
    kmeans[k] = sklearn.cluster.KMeans(n_clusters=k)
    kmeans[k].fit(ElectionsDF_NoAbs)
    results[k] = kmeans[k].labels_
    print "%0d: score %f" %(k, kmeans[k].score(ElectionsDF_NoAbs))
```

```python
results[10]
```

```python
results[5]
```

```python
#graph in 2 dimensions using PCA

PCA2 = sklearn.decomposition.PCA(n_components=2)
PCA2.fit(ElectionsDF_NoAbs)
```

```python
#Data frame of PCA components and kmeans labels
ResultsDF_NoAbs = pd.DataFrame(PCA2.transform(ElectionsDF_NoAbs)).join(pd.DataFrame(results))
ResultsDF_NoAbs.index = ElectionsDF_NoAbs.index
#Create column names so sns.lmplot will work
ResultsDF_NoAbs.rename(columns = lambda n: 'x'+str(n) if n < 2 else 'k'+str(n), inplace=True)
ResultsDF_NoAbs.head()
```

```python
#Create extra levels so merge will work
ResultsDF_NoAbs.columns = pd.MultiIndex.from_product([ResultsDF_NoAbs.columns, [''], ['']])

ResultsDF_NoAbs.head(10)
```

```python
import seaborn as sns
```

```python
%matplotlib inline
#scatter plots with cluster labels
for k in range(2, 11):
    sns.lmplot(data=ResultsDF_NoAbs, x='x0', y='x1', hue='k'+str(k), size=10, fit_reg=False)
```

```python
#Let us start with the 10-cluster version and see how the clusters differ.

Elections_NoAbs_Clusters = ElectionsDF_NoAbs.merge(ResultsDF_NoAbs, left_index=True, right_index=True, 
                                                   how='left')
Elections_NoAbs_Clusters.head(10)
```

Let us work to examine the nature of each cluster. We will look at the average by cluster of various variables related to the elections. If we see no significant differences between certain clusters, we may decide a model with fewer than 10 clusters is more appropriate.

```python
Elections_NoAbs_Means_k10 = Elections_NoAbs_Clusters.groupby('k10').mean()
Elections_NoAbs_Means_k10.columns
```
